## Loops (for/while)

```python
# Create a list of 5 numbers and use a for loop to compute the sum.
numbers = [8,18,28,38,48]
total = 0
for num in numbers:
  total += num
print("Sum:",total)
```

```python
# Loop through a dictionary and print each key and value.
cars = {"Brand" : "Toyota","Model" : "Corolla","Year": 2025,"Color": "Pink"}
for key,value in cars.items():
  print(f"{key} : {value}")
```

```python
# Loop through a list of names and print each one.
names = ["Nicole","Christian","Richie","Sponge Bob","Squidward","Peter","Luke","Gabby"]
for name in names:
  print(name)
```

```python
# Use a while loop to print numbers from 5 down to 1.
count = 5
while count >= 1:
  print(count)
  count -= 1
```

```python
# define the count as starting point, but it has to be greater than or equal to 1 cant go below 1 essentialy
count = 8
while count >= 1:
  print(count)
  count -= 1
```

```python
# Use a for loop to print numbers from 1 to 10.
for i in range(1,11): # 11 includes 10 10 only goes up to 9 within range
  print(i) # 0,10 includes 0 - 10
```

## CONDITIONAL STATEMENTS (if / elif / else)

```python
# Ask the user for their age and print "Adult" if age ≥ 18.
age = int(input("Enter your age: "))
if age >=18 :
  print("Adult")
else:
  print("Minor")
```

```python
# Write a program that checks if a number is even or odd.
num = int(input("Enter a number: "))
if num%2 == 0:
  print("Even")
else:
  print("Odd")
```

```python
# If temp > 30, print "Hot", else print "Cool".
temp = float(input("Enter temperature: "))
if temp>30:
  print("Hot")
else:
  print("Cool")
```

```python
# Write an if/elif/else to check grade range: ≥ 80 → "A" ≥ 60 → "B" else → "C"
grade = int(input("Enter your grade: "))
if grade >=80:
  print("A")
elif grade>=60:
  print("B")
else:
  print("C")
```

```python
# Ask the user for a password and check if it matches "12345".
password = input("Enter your password: ")
if password == "12345":
  print("Password match")
else:
  print("Different paassword used!!")
```

## DICTIONARIES

```python
# Create a dictionary called student with keys: name, age, grade.
student = {"name": "Nicole","age": "20","grade" : "A"}
print(student)
```

```python
# Access the value of the "name" key in the dictionary.
print(student["name"])
```

```python
# Change the "age" value to a new number
student["age"] = 21
print(student["age"])
print(student)
```

```python
# Add a new key "gender" to the dictionary.
student["gender"] ="Female"
print(student)
```

```python
# Remove the "grade" key from the dictionary.
del student["grade"]
print(student)
```

```python
# Loop through the dictionary keys and print each one.
student = {"name": "Nicole","age": "20","grade" : "A"}
print(student)
for key in student.keys():
  print(key)
```

```python
# Loop through the dictionary values and print each one.
student = {"name": "Nicole","age": "20","grade" : "A"}
print(student)
for value in student.values():
  print(value)
```

```python
# Print all key-value pairs using .items().
for key,value in student.items():
  print(f"{key} -> {value}")
```

```python
# Use in to check if "age" exists in the dictionary.
print("age" in student)
```

```python
# Create a nested dictionary representing two students.
students = {"student1": {"Name": "Alice","age":28,"grade": "B"},
            "student2": {"Name": "Rick","age":22,"grade": "A"}}
print(students)
for student_id, info in students.items():
  print(student_id,":" ,info)
```

```python
# Create a nested dictionary representing two students.
students = {"student1": {"Name": "Alice","age":28,"grade": "B"},
            "student2": {"Name": "Rick","age":22,"grade": "A"}}
for student_id, info in students.items():
  print(student_id,":" ,info)
```

```python
students = {"student1": {"Name": "Alice","age":28,"grade": "B"},
            "student2": {"Name": "Rick","age":22,"grade": "A"}}
for student_id, details in students.items():
    print(student_id + ":")
    for key, value in details.items():
        print(" ", key, "=", value)
```
